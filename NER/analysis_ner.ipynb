LLAMA3

```python
import numpy as np
llama = np.load(f'NER_clean_LLAMA3.npy',allow_pickle=True)
len(llama)
```

```python
llama_true_labels = []
llama_predicted_lebels = []
for item in llama:
    llama_true_labels.append(item['label'])
    llama_predicted_lebels.append(item['response'][1])
```

```python
for i,t in enumerate(llama_true_labels):
    if t == 7 :
        llama_true_labels[i]=1
    elif t == 8 :
        llama_true_labels[i]=2
    elif t == 9 :
        llama_true_labels[i]=3
    elif t == 10 :
        llama_true_labels[i]=4
    elif t == 11 :
        llama_true_labels[i]=5
    elif t == 12 :
        llama_true_labels[i]=6
    elif t == 13 :
        llama_true_labels[i]=7
```

```python
unique_values= np.unique(llama_true_labels, return_inverse=True)
unique_values
```

```python
tags = ["O", "Event", "Facility", "LOC", "ORG", "PER", "PRO"]
```

```python
for i,t in enumerate(llama_predicted_lebels):
    if t == "O" :
        llama_predicted_lebels[i]=int(0)
    elif t == "0" :
        llama_predicted_lebels[i]=int(0)
    elif t == "Event" :
        llama_predicted_lebels[i]=int(1)
    elif t == "Facility" :
        llama_predicted_lebels[i]=int(2)
    elif t == "LOC" :
        llama_predicted_lebels[i]=int(3)
    elif t == "ORG" :
        llama_predicted_lebels[i]=int(4)
    elif t == "PER" :
        llama_predicted_lebels[i]=int(5)
    elif t == "Product" :
        llama_predicted_lebels[i]=int(6)
    else : 
        llama_predicted_lebels[i]=int(0)
    
```

```python
unique_values= np.unique(llama_predicted_lebels, return_inverse=True)
unique_values
```

```python
from sklearn.metrics import f1_score

# Calculate Macro F1-score
macro_f1 = f1_score(llama_true_labels, llama_predicted_lebels, average='macro')

print(f'Macro F1-score: {macro_f1}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(true_labels, predicted_labels , filename):
    cm = confusion_matrix(true_labels, predicted_labels)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False, 
                xticklabels=["O", "Event", "Facility", "LOC", "ORG", "PER", "PRO"], yticklabels=["O", "Event", "Facility", "LOC", "ORG", "PER", "PRO"])
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.title(filename)
    plt.savefig(filename)
    plt.show()

plot_confusion_matrix(llama_true_labels, llama_predicted_lebels , 'Confusion Matrix for llama')
```
